```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
from scipy.ndimage import gaussian_filter
```

```python
img = cv2.imread('./1.jpg',0)
plt.imshow(img)
plt.show()
```

```python
new_1 = gaussian_filter(img, sigma=1)
plt.imshow(new_1)
plt.show()
```

```python
new_2 = gaussian_filter(img, sigma=2)
plt.imshow(new_2)
plt.show()
```

```python
new =  new_1 - new_2
plt.imshow(new)
plt.show()
```
